```python
def outliers(df):
    df2 = df.copy()
    df_col = df.select_dtypes(include = 'number').columns
    for col in df_col:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3-q1
        lr = q1-1.5*iqr
        up = q3+1.5*iqr
        df2 = df[(df[col] >= lr) & (df[col] <= up) ]
    return df2
```
